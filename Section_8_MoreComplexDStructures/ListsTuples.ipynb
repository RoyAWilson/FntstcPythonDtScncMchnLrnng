### Challenge:
### Create a long list that shows if value is odd or even in a sequence 0 to 1000

```python
long_list = []
for i in range(0, 10001):
    long_list.append(i)
    if long_list[i] % 2 == 0:
        print(f'{long_list[i]} is even')
    else:
        print(f'{long_list[i]} is odd')


        
```

```python
long_list_2 = []
for i in range(0, 1001):
    if i % 2 == 0:
        long_list_2.append('Even')
    else:
        long_list_2.append('Odd')
print(long_list_2)
```
