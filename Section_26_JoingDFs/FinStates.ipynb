```python
# Combine the tables to show data needed

combined_data = BS.merge(IS, on=['symbol', 'date'], how='inner', suffixes=['', 'IS'])
```

#### Lecture 155 - Financial Ratios Calculations

```python
combined_data['ROE'] = combined_data.NI / combined_data.Equity
combined_data['ROA'] = combined_data.NI / combined_data.Assets

## Set ROE <= 0 as np.nan and do same for Assets
combined_data.loc[combined_data.Equity<=0,'ROE'] = np.nan
combined_data.loc[combined_data.Assets<=0,'ROA'] = np.nan
```

```python
combined_data.columns
```

```python
combined_data[['ROE', 'ROA']].describe()
```

```python
# Clean data to remove problems with infinity:

combined_data.replace(np.inf, np.nan)
combined_data[['ROE', 'ROA']].describe()
# Probably would have been better to remove zeros from the data before calculating.
```
