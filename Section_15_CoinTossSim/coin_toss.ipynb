```python
# Import required packages:

import random
import statistics as stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Write function to return random 0 or 1 for heads or tails:

def toss(n):
    return [random.randint(0, 1) for item in range(0, n)]
```

```python
# Check the mean of 10 tosses:

stats.mean(toss(10))
```

##### Repeat tosses x times in function

```python
def experiment(x, n_tosses):
    results = []
    for i in range(0, x):
        outcome = toss(n_tosses)
        avg = stats.mean(outcome)
        results.append(avg)
    return results
```

```python
experimental_result = experiment(100, 100)
```

```python
# Visualise the data

plt.hist(experimental_result)
plt.title('Coin Toss Averages')
plt.ylabel('Number of Tosses')
plt.xlabel('Percentage Heads')
plt.savefig(r'Plots/Hist100100.png')
```
